```python
#อิมพอทข้อมูลpandas เข้าเพื่อใช้ในการทำงานโดยให้ชื่อ pandas เป็น pd
import requests
import io
import pandas as pd
```

```python
#นำข้อมูลตารางของวันที่ 1-10 เข้า pandasสร้างตารางข้อมูลของแต่ละวันโดยใช้ชื่อ url1-10 
url1 = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports/04-01-2020.csv")
url2 = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports/04-02-2020.csv")
url3 = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports/04-03-2020.csv")
url4 = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports/04-04-2020.csv")
url5 = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports/04-05-2020.csv")
url6 = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports/04-06-2020.csv")
url7 = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports/04-07-2020.csv")
url8 = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports/04-08-2020.csv")
url9 = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports/04-09-2020.csv")
url10 = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports/04-10-2020.csv")
url = [url1, url2, url3, url4, url5, url6, url7, url8, url9, url10] #สร้างตัวแปรให้ชื่อ url เพื่อเชื่อมข้อมูลทั้ง 10 วันเพื่อให้ง่ายต่อการใช้งาน
covid = pd.concat(url) #ทำการสร้างตัวแปรโดยในที่นี้ให้ชื่อ covid  ทำการเชื่อมข้อมูลของ url 
covid #รัน covid ออกมาเป็นข้อมูลตารางทั้งหมด
```

```python
#ข้อที่ 1 นับจำนวน Record ทั้งหมด
covid = pd.concat(url) #ทำการสร้างตัวแปรโดยในที่นี้ให้ชื่อ covid  ทำการเชื่อมข้อมูลของ url 
covid.count() #count ใช้ในการนับจำนวนของข้อมูลทั้งหมดโดยอิงจากข้อมูลทั้ง 10 วันที่เชื่อมต่อกับ url
```

```python
#ข้อที่ 2 จำนวนผู้เสียชีวิต(Death) ระหว่างวันที่ 1 - 2 เมษายน 2563
url = [url1, url2] #สร้างตัวแปรชื่อ url ใหม่ให้เท่ากับข้อมูลของวันที่ 1 และ 2 
covid = pd.concat(url) 
covid = pd.DataFrame(covid, columns= ['Last_Update','Deaths']) #ทำการสร้างดาต้าเฟรมเพื่อให้แสดงผลเป็นคอลัมที่เราจะเลือกคือ Last_Update กับ Deaths
covid 
```

```python
#ข้อที่ 3 จำนวนผู้เสียชีวิต(Death) ประเทศไทย(Thailand) วันที่ 7 เมษายน 2563
url7 = pd.DataFrame(url7, columns= ['Country_Region','Last_Update','Deaths']) #ทำการสร้างดาต้าเฟรมโดยใช้อิงจาก url7 หรือข้อมูลของวันที่ 7 โดยใช้ชื่อคอลัมที่เราต้องการ
covid = url7.loc[(url7['Country_Region'] == 'Thailand')] #ทำการสร้างตัวแปร covid โดยใช้ url7 และล็อคข้อมมูลของ Country_Region ให้เป็นประเทศไทย 
covid
```

```python
#ข้อที่ 4 จำนวนผู้รักษาหาย(Recovered) ระหว่างวันที่ 1 - 2 เมษายน 2563
url = [url1,url2] #สร้างตัวแปรชื่อ url ใหม่ให้เท่ากับข้อมูลของวันที่ 1 และ 2 
covid = pd.concat(url) 
covid = pd.DataFrame(covid, columns= ['Last_Update','Recovered']) #ทำการสร้างดาต้าเฟรมเพื่อให้แสดงผลเป็นคอลัมที่เราจะเลือกคือ Last_Update กับ Recovered
covid
```

```python
#ข้อที่ 5 จำนวนผู้ติดเชื้อ(Confirmed) รัฐ Wisconsin (Province_State) ในวันที่ 10 เมษายน 2563
url10 = pd.DataFrame(url10, columns= ['Province_State','Last_Update','Confirmed']) #ทำการสร้างดาต้าเฟรมโดยใช้อิงจาก url10 หรือข้อมูลของวันที่ 10 โดยใช้ชื่อคอลัมที่เราต้องการ
covid = url10.loc[(url10['Province_State'] == 'Wisconsin')] #ทำการสร้างตัวแปร covid โดยใช้ url7 และล็อคข้อมมูลของ Province_State ให้เป็น Wisconsin
covid
```

```python
#ตั้งโจทย์เอง ข้อที่ 1 จงหาจำนวนผู้เสียชีวิตและจำนวนผู้ติดเชื้อของวันที่ 6 ที่มีจำนวนมากกว่า 4
url6 = pd.DataFrame(url6, columns= ['Last_Update','Deaths']) #ทำการสร้างดาต้าเฟรมโดยใช้อิงจาก url6 หรือข้อมูลของวันที่ 6 โดยใช้ชื่อคอลัมที่เราต้องการ
covid = url6.loc[url6['Deaths'] > 4] #ทำการล็อคค่าให้ Deaths มีค่าน้อยกว่า 4 เพื่อหาจำนวนผู้เสียชีวิตที่น้อยกกว่า 4 
covid
```

```python
#ตั้งโจทย์เอง ข้อที่ 2 แสดงรายชื่อ Admin2 ที่ชื่อ Ada หรือประเทศ US พร้อมกับจำนวนผู้เสียชีวิตของวันที่ 3
url3 = pd.DataFrame(url3, columns= ['Admin2','Country_Region','Deaths']) #ทำการสร้างดาต้าเฟรมโดยใช้อิงจาก url3 หรือข้อมูลของวันที่ 3 โดยใช้ชื่อคอลัมที่เราต้องการ
covid = url3.loc[(url3['Admin2'] == 'Ada') | (url3['Country_Region'] == 'US')] #ล็อคค่าให้ Admin2 ให้เท่ากับ Ada และใช้ | เพื่อให้สามารถเพิ่มล็อคข้อมูลอีก 1 ชุดได้โดยให้ Country_Region เท่ากับ US
covid
```

```python
#ตั้งโจทย์เอง ข้อที่ 3 จำนวนผู้เสียชีวิตและจำนวนผู้รักษาหายที่อยู่ในประเทศไทย(Thailand)โดยที่ผู้เสียชีวิตมากกว่า 5 คน
url1 = pd.DataFrame(url1, columns= ['Country_Region','Deaths','Recovered']) #ทำการสร้างดาต้าเฟรมโดยใช้อิงจาก url3 หรือข้อมูลของวันที่ 3 โดยใช้ชื่อคอลัมที่เราต้องการ
test = url1.loc[(url1['Country_Region'] == 'Thailand') | (url1['Deaths'] > 5)] ##ล็อคค่าให้ Country_Region ให้เท่ากับ Thailand และใช้ | เพื่อให้สามารถเพิ่มล็อคข้อมูลอีก 1 ชุดได้โดยให้ Deaths มากกว่า 5
test
```
